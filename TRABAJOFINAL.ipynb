```python
CORRELACIONES
Una matriz de correlación es una herramienta estadística que muestra la intensidad y la dirección de la relación entre dos o más variables.

La matriz de correlación nos permite, mediante coeficientes de correlación de distintas variables cómo se relacionan entre sí todos los posibles pares de valores de una tabla. Es una poderosa herramienta para resumir un gran conjunto de datos y encontrar y mostrar patrones en ellos.
```

```python
#Medición de la correlación entre las distintas variables
df.corr (numeric_only=True)
```

```python
fig = plt.figure(figsize = (20,5))
sns.heatmap(df.corr(numeric_only=True),annot=True, cmap="RdYlGn")
plt.show()
```

```python
Entendiendo que la correlación varía en un rango de 0 a 1, se puede observar:
- una correlación positiva del 91% entre el "índice de emprendimiento" con "Índice de emprendimiento de mujeres"
- una correlación positiva del 44% entre "Índice de emprendimiento de mujeres" con "Tasa de participacipon femenina en la fuerza laboral"
- una correlación negativa del 46% entre "Tasa de inflación" e "Índice de emprendimiento de mujeres"
- una correlación negativa del 14% entre "Tasa de inflación" con "Tasa de participacipon femenina en la fuerza laboral"
```
